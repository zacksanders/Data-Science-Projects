## Data Cleaning 

Here we take COVID data from every country and clean it up. Countries with negligible populations are thrown out. From the cumulative daily cases we obtain daily cases and a seven day rolling average. 

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt



# df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv", sep=",")
df = pd.read_csv("time_series_covid19_confirmed_global.txt", sep=",")
df = df.drop(['Province/State', 'Lat', 'Long'], axis=1)
df = df.rename(columns={'Country/Region' : 'Country'})
df = df.rename(index = {'US' : 'United States', 'Korea, South' : 'South Korea', })


df = df.groupby(["Country"]).sum()
# print(df.index.tolist())

dtf = pd.read_csv("population.csv", sep=',')
dtf = dtf[["name","pop2019"]]
dtf = dtf[dtf["pop2019"] > 1000]


df["sig_pop"] = df.index.isin(dtf.name.tolist())
df = df[df.sig_pop]
len(df.columns)

# print('\n')
# print(dtf['name'].tolist())

country = "Thailand"
daily = [df.loc[country][i] - df.loc[country][i-1] for i in range(len(df.loc[country]))]

max = np.max(daily)

N=7

sev_day_avg = np.convolve(daily, np.ones((N,))/N, mode='valid')
# pd.Series(daily).rolling(window=N).mean().iloc[N-1:].values


y = df.loc[country]
x = list(range(len(y)))

y = y[:-1]
x = x[:-1]
daily = daily[:-1]
sev_day_avg = sev_day_avg[:-1]

d = daily
av = sev_day_avg

def extractDigits(lst): 
    return [[el] for el in lst]
x = extractDigits(x)

plt.figure(figsize=(15,4))

plt.subplot(1, 3, 1)
plot = plt.plot(x,y)
plt.title('Cumulative Cases')
plt.xlabel('days')
plt.ylabel('cases')

plt.subplot(1, 3, 2)
plot = plt.plot(x,d)
plt.axis([0, 137, 0, max+300])
plt.title('Daily Cases')
plt.xlabel('days')
plt.ylabel('cases')

x = x[:-(N-1)]
plt.subplot(1, 3, 3)
plot = plt.plot(x,av)
plt.axis([0, 137, 0, max+300])
plt.title('Daily Cases (Seven Day Rolling Average)')
plt.xlabel('days')
plt.ylabel('cases')

print(y[-2:-1])
print(x[-2:-1])

plt.subplots_adjust(wspace=0.4)

plt.show()
```
